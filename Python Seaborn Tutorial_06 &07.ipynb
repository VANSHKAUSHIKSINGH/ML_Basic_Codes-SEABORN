# How To Draw Seaborn Barplot / Bar Graph

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips_df = sns.load_dataset("tips")
tips_df
```

```python
sns.barplot()
```

```python
sns.barplot(x = tips_df.total_bill)
```

```python
sns.barplot(x = tips_df.day,y = tips_df.total_bill)
```

```python
sns.barplot(x = tips_df.day,y = tips_df.total_bill,hue = tips_df.sex)
```

```python
sns.barplot(x = 'day', y = 'total_bill', hue ='sex', data =  tips_df)
```

# How To Draw Seaborn Barplot / Bar Graph #2

```python
kwarg = {'alpha':0.9, 'linestyle':':','linewidth':5, 'edgecolor':'k'}

sns.barplot(x = 'day', y = 'total_bill',hue = 'sex',data = tips_df,**kwarg)
```

```python
sns.set()
kwarg = {'alpha':0.9, 'linestyle':':','linewidth':5, 'edgecolor':'k'}

sns.barplot(x = 'day', y = 'total_bill',hue = 'sex',data = tips_df,**kwarg)
```

```python
sns.set()
kwarg = {'alpha':0.9, 'linestyle':':','linewidth':5, 'edgecolor':'k'}

sns.barplot(x = 'day', y = 'total_bill',data = tips_df,**kwarg)
```

```python
sns.barplot(x = 'day',y = 'total_bill',data = tips_df,alpha= .5, linestyle = '-.',linewidth = 6,edgecolor = 'g')
```

```python
ax  = sns.barplot(x = 'day',y = 'total_bill',data= tips_df, alpha = .9,linestyle = '-.',linewidth = 6,edgecolor = 'g')
ax.set(title = "Barplot of Tips DataFrame", xlabel = 'Days', ylabel = "Total_Bill")
```

```python
plt.figure(figsize = (16,9))
sns.barplot(x = 'day',y = 'total_bill',data= tips_df, alpha = .9,linestyle = '-.',linewidth = 3,edgecolor = 'g')

plt.title("Barplot of Day and Total_bill", fontsize = 20)
plt.xlabel("Days",fontsize = 18)
plt.ylabel('Total Bill',fontsize = 15)

plt.savefig("Barplot of Day and Total_bill")
plt.show()
```
