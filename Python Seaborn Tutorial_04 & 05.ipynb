# How To Draw Seaborn Histogram / Seaborn Distplot

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm
```

```python
tips_df= sns.load_dataset("tips")
tips_df
```

```python
sns.displot(tips_df["size"])
```

```python
sns.displot(tips_df["tip"])
```

```python
sns.displot(tips_df["total_bill"])
```

```python
sns.displot(tips_df["total_bill"],bins = 100)
```

```python
sns.distplot(tips_df["total_bill"], hist = False,)
```

```python
sns.distplot(tips_df["total_bill"],kde = False,)
```

```python
sns.distplot(tips_df["total_bill"],rug = True,)
```

```python
sns.distplot(tips_df["total_bill"],fit =norm,)
```

```python
sns.distplot(tips_df["total_bill"],fit =norm,kde = False)
```

```python
sns.distplot(tips_df["total_bill"],label = "Total Bill",)
plt.title("Histogram of Total Bill")
plt.legend()
```

```python
sns.set()
sns.distplot(tips_df["total_bill"],label = "Total Bill")
plt.title("Histogram of Total Bill")
plt.legend()
```

```python
plt.figure(figsize= (16,9))
sns.set()
sns.distplot(tips_df["total_bill"],label = "Total Bill")
plt.title("Histogram of Total Bill")
plt.legend()
```

```python
tips_df.total_bill.sort_values()
```

```python
bins = [1,5,10,15,20,25,30,35,40,45,50,55]

plt.figure(figsize = (16,9))
sns.set()

sns.distplot(tips_df["total_bill"], bins = bins)

plt.xticks(bins)

plt.title("Histogram of Total Bill")
plt.show()
```

```python
plt.figure(figsize = (16,9))
sns.set()

sns.distplot(tips_df["total_bill"], hist_kws = {'color':'#DC143C','edgecolor':'#aaff00',})
```

```python
plt.figure(figsize = (16,9))
sns.set()

sns.distplot(tips_df["total_bill"], hist_kws = {'color':'#DC143C','edgecolor':'#aaff00','linewidth':5,})
```

```python
plt.figure(figsize = (16,9))
sns.set()

sns.distplot(tips_df["total_bill"], hist_kws = {'color':'#DC143C','edgecolor':'#aaff00','linewidth':5,'linestyle':'--',})
```

```python
plt.figure(figsize = (16,9))
sns.set()

sns.distplot(tips_df["total_bill"], hist_kws = {'color':'#DC143C','edgecolor':'#aaff00','linewidth':5,'linestyle':'--',
                                               'alpha':0.9,})
```

```python
plt.figure(figsize = (16,9))
sns.set()

sns.distplot(tips_df["total_bill"], 
             hist_kws = {'color':'#DC143C','edgecolor':'#aaff00','linewidth':5,'linestyle':'--',
                                               'alpha':0.9,},
             kde_kws = {'color':'#8e00ce','linewidth':5,'linestyle':'--',
                                               'alpha':0.9,},
            rug = True)
```

```python
plt.figure(figsize = (16,9))
sns.set()

sns.distplot(tips_df["total_bill"], 
             hist_kws = {'color':'#DC143C','edgecolor':'#aaff00','linewidth':5,'linestyle':'--',
                                               'alpha':0.9,},
             kde_kws = {'color':'#8e00ce','linewidth':5,'linestyle':'--',
                                               'alpha':0.9,},
            rug = True,
            rug_kws = {'color':'#0426d0','linewidth':5, 
                       'linestyle':'--','alpha':0.9},)
```

```python
plt.figure(figsize= (16,9))
sns.set

sns.distplot(tips_df["total_bill"],
            hist_kws = {'color':'#DC143C','edgecolor':'#aaff00','linewidth':5,'linestyle':'--','alpha':0.9},
            
            kde = False,
            fit  = norm,
            fit_kws = {'color':'#8e00Ce','linewidth':12,'linestyle':'--','alpha':0.6},
            
            rug = True,
            rug_kws ={'color':'#0426d0','linewidth':5,'linestyle':'--','alpha':0.4},
             label = "TB")
plt.title("Histogram of Restorant Total Bill", fontsize=20)
plt.xlabel("Total Bill", fontsize = 15)
plt.ylabel("Density", fontsize = 25)
plt.legend()
plt.show()
```

```python
plt.figure(figsize = (16,9))
sns.distplot(tips_df["total_bill"], bins = 9, label = "total_bill")
sns.distplot(tips_df["tip"], bins = 9, label = "tip")
sns.distplot(tips_df["size"], bins = 9, label = "size")

plt.legend()
```
